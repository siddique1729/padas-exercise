```python
# To divide the Wind Direction into 4 quadrants
# [0 - 90) (N-E); [90 - 180) (E-S); [180-270) (S-W) and [270 - 360)
# Label as ‘North’, ‘East’, ‘South’ and ‘West’
```

```python
# Extracting the wind direction column as pandas series
wdir = a['WDIR.ts']
wdir = list(wdir) # converting to list

col_len = len(wdir)

# converting the list elements to int

for i in range(0,col_len,1):
    wdir[i] = float(wdir[i])
```

```python
len(wdir)
```

```python
wcat = [] # creating an empty list of strings for wdir category
j = 0 # counter for the list

# Creating a list of wdir category
for i in range(0,col_len,1):
    if wdir[i] >= 0 and wdir[i] < 90:
        wcat.append('North')
    elif wdir[i] >= 90 and wdir[i] < 180:
        wcat.append('East')
    elif wdir[i] >= 180 and wdir[i] < 270:
        wcat.append('South')
    else:
        wcat.append('West')

# converting the wcat list to pandas series

wdircat = pd.Series(wcat)
wdircat

# Adding the series to the dataframe

#new_hr = a.assign(candidates = cand_s)

a['Wind_Direction'] = wdircat
a
```

```python
# To compute the minimum. mean, standard deviation and maximum windspeeds for winds from the 'North'.
```

```python
# Extracting the dataframe for wind direction 'North'

a2 = a.loc[(a['Wind_Direction'] == 'North')]
a2.head()

# Extracting the wind speed column as pandas series
wspeed = a['WSPEED.ts']
wspeed = list(wspeed) # converting to list

col_len = len(wspeed)

# converting the list elements to int

for i in range(0,col_len,1):
    wspeed[i] = float(wspeed[i])

wspeed
```

```python
# Computing minimum, maximum & mean

# Dummy values for min, max, sum

summ = 0
minn = 10000000000000000000000
maxx = -1

for i in range(0,col_len,1):
    if wspeed[i]<minn:
        minn = wspeed[i]
    if wspeed[i]>maxx:
        maxx = wspeed[i]
    
    summ = summ + wspeed[i]
    
avg = summ/col_len

print(minn,maxx,avg)
print(min(wspeed),max(wspeed),sum(wspeed)/col_len) # values check

# Computing standard deviation
# Formula - sqrt((sum(xi-avg)*(xi-avg))/N)

summm = 0
for i in range(0,col_len,1):
    summm = summm + (wspeed[i]-avg)*(wspeed[i]-avg)
    
std = mt.sqrt(summm/col_len)
print(std)
print(st.stdev(wspeed)) # values check
```

```python
# While loop - to compute the average ozone concentration when the temperature is greater than 70F.
```

```python
# Extracting the dataframe for temperature > 70F

a2 = a.loc[(a['TEMP.ts'] > 70)]
a2.head()

# Extracting the ozone concentration as pandas series
oz = a2['OZONE.ts']
oz = list(oz) # converting to list

col_len = len(oz)

# converting the list elements to int

for i in range(0,col_len,1):
    oz[i] = float(oz[i])

oz

i = 0
summ = 0

while i < col_len:
    summ = summ + oz[i]
    i = i + 1

oz_avg = summ/col_len
print(oz_avg)
```

```python
# For loop - to compute the standard deviation of the ozone when temperature is > 70F AND NOX > 10 ppb
```

```python
# Extracting the dataframe for temperature > 70F

a2 = a.loc[(a['TEMP.ts'] > 70) & (a['NOX.ts'] > 10)]
a2.head()

# Extracting the ozone concentration as pandas series
oz = a2['OZONE.ts']
oz = list(oz) # converting to list

col_len = len(oz)

# converting the list elements to int

for i in range(0,col_len,1):
    oz[i] = float(oz[i])

oz

i = 0
summ = 0

while i < col_len:
    summ = summ + oz[i]
    i = i + 1

oz_avg = summ/col_len

# Computing standard deviation
# Formula - sqrt((sum(xi-avg)*(xi-avg))/N)

summm = 0
for i in range(0,col_len,1):
    summm = summm + (oz[i]-oz_avg)*(oz[i]-oz_avg)
    
std = mt.sqrt(summm/col_len)
print(std)
print(st.stdev(oz)) # values check
```
